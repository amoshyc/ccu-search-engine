```python
%matplotlib inline
import json
import pathlib
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('seaborn')
from tqdm import tqdm
```

```python
json_paths = list(pathlib.Path('./data/').glob('*.json'))
cnt_ingr = np.zeros(len(json_paths), dtype=np.int32)
cnt_step = np.zeros(len(json_paths), dtype=np.int32)
for i, path in enumerate(tqdm(json_paths)):
    with path.open() as f:
        data = json.load(f)
    cnt_ingr[i] = data['n_ingr']
    cnt_step[i] = data['n_step']
```

```python
fig, ax = plt.subplots(dpi=100, figsize=(15, 8))
ax.hist(cnt_ingr, 100)
ax.set_xlabel('#Ingredients')
ax.set_ylabel('Number of Recipes')
plt.show()
```

```python
fig, ax = plt.subplots(dpi=100, figsize=(15, 8))
ax.hist(cnt_step, 30)
ax.set_xlabel('#Steps')
ax.set_ylabel('Number of Recipes')
plt.show()
```

```python
print(np.mean(cnt_ingr))
print(np.mean(cnt_step))
```
